```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
```

```python
cancer = load_breast_cancer()
X = cancer.data[:,:2]
y = cancer.target
```

```python
plt.scatter(X[:, 0], X[:, 1], 
            c=y, 
            s=20, edgecolors="k")
```

```python
cancer.feature_names
```

```python
cancer.target_names
```

```python
df = load_breast_cancer(return_X_y = True, as_frame = True)
```

```python
svm = SVC(kernel = 'rbf' , gamma = 0.5, C=10000.0)
svm.fit(X, y)
```

```python
DecisionBoundaryDisplay.from_estimator(svm,
                                       X,
                                       response_method = "predict",
                                       cmap = plt.cm.Spectral,
                                       alpha = 0.8,
                                       xlabel = cancer.feature_names[0],
                                       ylable = cancer.feature_names[1]
                                      )
plt.scatter(X[:, 0], X[:, 1], 
            c=y, 
            s=20, edgecolors="k")
plt.scatter(svm.support_vectors_.T[0],svm.support_vectors_.T[1], c = "r")
plt.show()
```
