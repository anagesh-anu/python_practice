```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

## Creating test data using make_blobs 

```python
#centroids = [(1,1)]

X, y = make_blobs(n_samples=100, 
                   n_features=2, 
                   centers=4, 
                   #cluster_std=cluster_std, 
                   #center_box=(-10.0, 10.0),
                     #random_state = 1
                 )
```

```python
X
```

```python
y
```

### Created test data have the below clusters

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1])
```

## Trying elbow method to find out the optimum number of clusters

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,10):
    km = KMeans(n_clusters=i)
    km.fit_predict(X)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,10), wcss)
plt.show()
```

### As per the above graph '4' is the optimum number for the clusters

## So Initializing KMeans with 4 number of clusters

```python
km = KMeans(n_clusters = 4)
y_pred = km.fit_predict(X)
```

```python
y_pred
```

```python
plt.scatter(X[y_pred == 0,0],X[y_pred == 0,1],color='blue')
plt.scatter(X[y_pred == 1,0],X[y_pred == 1,1],color='red')
plt.scatter(X[y_pred == 2,0],X[y_pred == 2,1],color='green')
plt.scatter(X[y_pred == 3,0],X[y_pred == 3,1],color='yellow')
#plt.scatter(X[y_pred == 4,0],X[y_pred == 4,1],color='black')
```
